```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import files
nvs2_total = "Resources/NVS2_total_total.csv"
nvs2_total_df = pd.read_csv(nvs2_total)

nvs2_nights = "Resources/NVS2_total_nights.csv"
nvs2_nights_df = pd.read_csv(nvs2_nights)

nvs2_avg = "Resources/NVS2_total_avg.csv"
nvs2_avg_df = pd.read_csv(nvs2_avg)

nsw_daytrip = "Resources/NSW_daytrips.csv"
nsw_daytrip_df = pd.read_csv(nsw_daytrip)

wa_daytrip = "Resources/WA_daytrips.csv"
wa_daytrip_df = pd.read_csv(wa_daytrip)

perth_month21 = "Resources/Perth by month.21-22.csv"
perth_month21_df = pd.read_csv(perth_month21)

sydney_month21 = "Resources/Sydney by month.21-22.csv"
sydney_month21_df = pd.read_csv(sydney_month21)

perth_month19 = "Resources/Perth by month.19-22.csv"
perth_month19_df = pd.read_csv(perth_month19)

sydney_month19 = "Resources/Sydney by month.19-22.csv"
sydney_month19_df = pd.read_csv(sydney_month19)
```

```python
list1 = nvs2_total_df['Unnamed: 0']
```

```python
#select the columns with the relevant years
nvs2_total_clean = nvs2_total_df[['Year ending June 2018', 'Year ending June 2019', 'Year ending June 2020', 'Year ending June 2021', 'Year ending June 2022']].copy()
nvs2_nights_clean = nvs2_nights_df[['Year ending June 2018', 'Year ending June 2019', 'Year ending June 2020', 'Year ending June 2021', 'Year ending June 2022']].copy()
nvs2_avg_clean = nvs2_avg_df[['Year ending June 2018', 'Year ending June 2019', 'Year ending June 2020', 'Year ending June 2021', 'Year ending June 2022']].copy()
nsw_daytrip_clean = nsw_daytrip_df[['Daytrips (000)', 'Year ending June 2018', 'Year ending June 2019', 'Year ending June 2020', 'Year ending June 2021', 'Year ending June 2022']].copy()
wa_daytrip_clean = wa_daytrip_df[['Daytrips', 'Year ending June 2018', 'Year ending June 2019', 'Year ending June 2020', 'Year ending June 2021', 'Year ending June 2022']].copy()
```

```python
#transpose the dataframes
nvs2_total_clean = nvs2_total_clean.T
nvs2_avg_clean = nvs2_avg_clean.T
nvs2_nights_clean = nvs2_nights_clean.T
```
