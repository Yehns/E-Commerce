```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import file
file1 = "Resources/tourism_output_per_industry.csv"
file2 = "Resources/tourism_output_per_related_industry.csv"
```

```python
#Extract file highlighting spend per tourism industry: 
output_per_industry_df= pd.read_csv(file1)
output_per_industry_df.info()
#Remove commas 
for col in output_per_industry_df:
    output_per_industry_df[col] = output_per_industry_df[col].str.replace(',','')
    #convert columns to integers
    if output_per_industry_df[col].name !='Tourism characteristic industries':
        output_per_industry_df[col] =pd.to_numeric(output_per_industry_df[col])    
output_per_industry_df.info()
output_per_industry_df.head()
```

```python
#Create a series for accommodation spent

accommodation_spend = output_per_industry_df.iloc[0,1:6]
print(accommodation_spend)
print(type(accommodation_spend))
```

```python
accommodation_spend.plot()
plt.savefig("Accommodation_spend.png")
```

```python
#Create a series for cafes and restaurants spend

restaurant_spend = output_per_industry_df.iloc[2,1:6]
print(restaurant_spend)
print(type(restaurant_spend))
```

```python
restaurant_spend.plot()
plt.savefig("Restuarant_spend.png")
```

```python
#Create a series for rail transport spend

rail_transport_spend = output_per_industry_df.iloc[4,1:6]
print(rail_transport_spend)
print(type(rail_transport_spend))
```

```python
plt.plot(rail_transport_spend)
plt.savefig("Rail_spend.png")
```

```python
#Combine plots
#x_axis = np.arange(len(output_per_industry_df))
x_axis = ("2017-18","2018-19","2019-20","2020-21","2021-22")
fig, ax = plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(8)
ax.plot(x_axis, (output_per_industry_df.iloc[0,1:6]), label=(output_per_industry_df.iloc[0,0]))
ax.plot(x_axis, (output_per_industry_df.iloc[1,1:6]), label=(output_per_industry_df.iloc[1,0]))
ax.plot(x_axis, (output_per_industry_df.iloc[2,1:6]), label=(output_per_industry_df.iloc[2,0]))
ax.plot(x_axis, (output_per_industry_df.iloc[3,1:6]), label=(output_per_industry_df.iloc[3,0]))
ax.plot(x_axis, (output_per_industry_df.iloc[4,1:6]), label=(output_per_industry_df.iloc[4,0]))
ax.plot(x_axis, (output_per_industry_df.iloc[5,1:6]), label=(output_per_industry_df.iloc[5,0]))
plt.xlabel('Time Period Tracked')
plt.ylabel('AUD Spend')
plt.title('Spend Pre- and Post-COVID in Key Tourism Industries')
plt.savefig("Industry_spend_analysis.png")
plt.show()
```
