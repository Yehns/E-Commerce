```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import file
file3 = "tourism_jobs_created_per_industry.csv"
```

```python
#Extract file highlighting spend per tourism industry:
jobs_per_industry_df = pd.read_csv(file3)
jobs_per_industry_df.info()
```

```python
#Confirm Data Frame
jobs_per_industry_df.head(16)
```

```python
#Reduce Data Frame to remove NaN factors
jobs_data_df = jobs_per_industry_df.iloc[0:13,0:7]
jobs_data_df.head(15)
```

```python
#Create bar graph
label = jobs_data_df['Tourism part and full-time jobs created (per 1000)']
x_axis = jobs_data_df['2018-19']
pre_covid_totals = jobs_data_df['2018-19']
post_Covid_totals = jobs_data_df['2021-22']

x = np.arange(len(x_axis))  
width = 0.35 

#Change format of graph to horizontal graph
fig, ax = plt.subplots()
rects1 = ax.barh(x - width/2, pre_covid_totals, width, label='2018-19')
rects2 = ax.barh(x + width/2, post_Covid_totals, width, label='2021-22')
fig.set_figwidth(10)
fig.set_figheight(10)

#Add some text for labels, title and custom y-axis tick labels, etc.
ax.set_ylabel('Job Types')
ax.set_xlabel('Number of Jobs (per 1,000)')
ax.set_title('Tourism Affiliated Job Numbers')
ax.set_yticks(x)
ax.set_yticklabels(label)
ax.legend()


fig.tight_layout()
plt.savefig("Impact_on_tourism_job_count.png")

plt.show()
```
